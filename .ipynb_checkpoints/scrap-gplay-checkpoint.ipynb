```python
from google_play_scraper import Sort, reviews
import pandas as pd
import numpy as np
import os

def extract_myim3_reviews(count=10000):
    print("Menginisialisasi koneksi ke server Play Store...")
    
    # Menarik ulasan dengan filter relevansi waktu
    result, continuation_token = reviews(
        'com.pure.indosat.care',
        lang='id', 
        country='id', 
        sort=Sort.NEWEST, 
        count=count 
    )
    
    # Transformasi JSON ke struktur tabular
    df = pd.DataFrame(np.array(result),columns=['review'])
    df = df.join(pd.DataFrame(df.pop('review').tolist()))
    
    # Filter kolom esensial untuk NLP
    df_filtered = df[['reviewId', 'userName', 'content', 'score', 'at', 'replyContent']]
    
    # Pembuatan direktori dan ekspor
    os.makedirs('dataset_indosat', exist_ok=True)
    file_path = 'dataset_indosat/myim3_reviews_raw.csv'
    df_filtered.to_csv(file_path, index=False)
    
    print(f"[SUCCESS] Ekstraksi selesai. {len(df_filtered)} baris disimpan di {file_path}")
    return df_filtered

if __name__ == "__main__":
    df_reviews = extract_myim3_reviews()
```
