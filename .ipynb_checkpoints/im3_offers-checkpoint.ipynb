```python
import json
import pandas as pd
import os

def parse_im3_offers(json_filepath):
    """
    Mengekstrak arsitektur harga dari JSON IOH dan mengubahnya menjadi DataFrame.
    """
    with open(json_filepath, 'r') as file:
        data = json.load(file)

    # Validasi struktur respons
    if data.get('status') != "0" or 'commercial_package' not in data.get('data', {}):
        raise ValueError("Format JSON tidak dikenali atau request gagal.")

    packages = data['data']['commercial_package']
    extracted_data = []

    for pkg in packages:
        # Ekstraksi atribut komersial bertingkat
        commercial_attr = pkg.get('commercial_attribute', {})
        
        # Penanganan tipe data untuk kuota (sering kali string numerik)
        try:
            total_quota_mb = float(commercial_attr.get('total_data_quota', 0))
            quota_gb = round(total_quota_mb / 1024, 2)
        except ValueError:
            quota_gb = 0

        # Ekstraksi komponen paket dan VAS dari subtitle menggunakan regex atau pemisahan string sederhana
        subtitle = pkg.get('package_subtitle', '')
        
        extracted_data.append({
            'offer_id': pkg.get('pvr_code'),
            'package_name': pkg.get('package_name'),
            'family_name': commercial_attr.get('family_name'),
            'quota_gb': quota_gb,
            'duration_days': commercial_attr.get('duration_month'),
            'price_rp': pkg.get('tariff'),
            'original_price_rp': pkg.get('original_tariff'),
            'marketing_ribbon': pkg.get('package_ribbon'),
            'vas_included': pkg.get('package_footer_desc'),
            'auto_renew': commercial_attr.get('auto_renew')
        })

    df = pd.DataFrame(extracted_data)
    
    # Hitung metrik Harga per GB
    df['price_per_gb'] = round(df['price_rp'] / df['quota_gb'], 2)
    
    return df

# Asumsi kamu menyimpan respons raw JSON di direktori ini
json_file = 'dataset_indosat/im3_offers_raw.json'
csv_output = 'dataset_indosat/im3_pricing_architecture.csv'

# Hanya jalankan jika file eksis untuk menghindari error di IDE
if os.path.exists(json_file):
    df_pricing = parse_im3_offers(json_file)
    df_pricing.to_csv(csv_output, index=False)
    print(f"[SUCCESS] Ekstraksi {len(df_pricing)} paket. Tersimpan di {csv_output}")
    print(df_pricing[['package_name', 'quota_gb', 'price_rp', 'price_per_gb']].head())
```
